```python
possible_errors = [0,1,2,3,5]
def create_observation(failed_components: Dict[str, int]):
    # assumption: observation of failed components can not differ from the real state
    # "we cant unfail components"
    all_failed_components = failed_components
    while len(failed_components.keys()) > 0:
        new_failed_components = {}
        for component, errorId in failed_components.items():
            if component in all_failed_components:
                continue
            else:
                probabilities = matrix[component][errorId]
                for computed_component in matrix.columns:
                    new_state = numpy.random.choice(possible_errors, p=probabilities[computed_component])
                    if new_state == 0:
                        continue
                    else:
                        all_failed_components[computed_component] = new_state
                        new_failed_components[computed_component] = new_state
        failed_components = new_failed_components
        
    return all_failed_components



```
